```python
from scipy.spatial.transform import Rotation
import numpy as np
import mpmath as mp
from orbit3d import plot_3dtraj, _rotation_matrix, _rotate, get_3dtrajectory, get_min_approach3d
from orbit2d import get_2dtrajectory
mp.mp.dps = 50
```

```python
r0 = mp.matrix([1e20, 6e3, 0])
v0 = mp.matrix([-200, 0., 0])
k = 1.32712e11
```

```python
get_min_approach3d(r0, v0, k)
```

```python
(thetamin, thetamax), r, v, t = get_3dtrajectory(r0, v0, k)
thetas = mp.linspace(thetamin, thetamax, 10_000)[1:-1]
```

```python
rs = [r(theta) for theta in thetas]
rs
```

```python
ts = [t(theta) for theta in thetas]
ts
```

```python
fig = plot_3dtraj(r0, v0, k)
fig.update_layout(
    scene = dict(
        xaxis = dict(range=[-100,100],),
        yaxis = dict(range=[-100,100],),
        zaxis = dict(range=[-100,100],)
    )
)
```
