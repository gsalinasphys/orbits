```python
from scipy.spatial.transform import Rotation
import numpy as np
import mpmath as mp
from orbit3d import get_3dtrajectory, plot_3dtraj
mp.mp.dps = 50
```

```python
rin, vin = np.array([-1.e20, -6e3, 0.]), np.array([-200., 0., 0.])
```

```python
k = 1.32712e11
theta0, xs, vs, ts = get_3dtrajectory(rin, vin, k)
```

```python
thetas = mp.linspace(0., 2*theta0, 100)
```

```python
xs(thetas)
```

```python
ts(thetas)
```
